#### 인스티즈 사이트 댓글 크롤링 작업

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from time import *

driver = webdriver.Chrome()
driver.get('https://www.instiz.net/pt/7276279')

comments = []

for a in range(16):
    html = driver.page_source
    # print(html)
    soup = BeautifulSoup(html, 'html.parser')

    reviews = soup.find_all('div', {'class':'comment_line'})
    
    for review in reviews:
      extract = soup.find('span', attrs={'class':'minitext'}).extract()
      print(review.get_text().strip())
      print('----------------')
      comments.append(review.get_text().strip())
    
    sleep(3)
```

```python
comments
```

```python
full_comment = pd.DataFrame(comments)
full_comment

full_comment.to_csv('full_comment.csv', encoding='utf-8-sig')
# 크롤링 작업한 원본 데이터 저장
```

```python
# 처리 작업 테스트
import re

com = comments[0]
com = com.split('\n')
print(com)
if re.match("^[a-zA-Z]+$", com[0]):
    print('alphablet')
else:
    print('not alpha')
```

```python
###########################################################################
# 문단별로 나누었을 때 첫 문장(mbti 기제)이 영어인 경우만 추출하여 데이터로 사용하기
# 사람들이 적은 mbti에 s n을 구분하여 column으로 지정
# 질문 1, 2, 3에 대한 답변을 따로 구분하여 column으로 지정

import re

user_comment = []

for comment in comments:
    # print(comment)
    # print(comment.split('\n'))
    comment_list = comment.split('\n')
    if re.match("^[a-zA-Z]+$", comment_list[0]):
        print(comment_list)
        print('------------')
        user_comment.append(comment_list)
```

```python
n_cnt, s_cnt = 0,0
for com in user_comment:
    if 'n' or 'N' in com[0]:
        n_cnt += 1
    elif 's' or 'S' in com[0]:
        s_cnt += 1
print(n_cnt, s_cnt)
```

```python
# s n 구분 테스트 
user_com = user_comment[7]
print(user_com)

user_com = list(filter(None, user_com))
# us_com = ''.join(user_com).split()

if 'n' or 'N' in user_com[0]:
    print('mbti : n')
    user_com[0] = 'N'
    print(user_com)
else:
    print('mbti : s') 
```

```python
mbti_comment = []

for us_com in user_comment:
    if len(us_com) == 4:
        # print(us_com)
        if 'n' or 'N' in us_com[0]:
            us_com[0] = 'N'
            mbti_comment.append(us_com)
        elif 's' or 'S' in us_com[0]:
            us_com[0] = 'S'
            mbti_comment.append(us_com)

user_comment
```
